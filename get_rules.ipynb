```python
from huggingface_hub import InferenceClient
```

```python
model_name = "Qwen/Qwen2.5-72B-Instruct"
```

```python
client = InferenceClient(model_name, token='')
```

```python
framework = 'lexical-functional grammar'
```

```python
framework_description = client.chat.completions.create(
        max_tokens=3000,
        messages=[
            {
                "role": "system",
                "content": "Describe a linguistic framework in a formalized way.\n"
                           "Return a list of rules in a form of an ordered list."
            },
            {
                "role": "user",
                "content": f"Describe the following framework: {framework}"
            }
        ]
    ).choices[0].message.content

print(framework_description)
```

```python
import ast

bnf_description = ast.literal_eval(client.chat.completions.create(
        max_tokens=5000,
        messages=[
            {
                "role": "system",
                "content": "Generate a BNF description based on the provided set of rules."
            },
            {
                "role": "user",
                "content": f"Generate BNF using the following rules: {framework_description}"
            }
        ],
        response_format={
            "type": "json",
            "value": {
                "properties": {
                    "bnf_desciption": {"type": "array", "items": {"type": "string"}}
                }
            }
        }
    ).choices[0].message.content)['bnf_desciption']

for rule in bnf_description:
  print(rule)
```

```python
with open('bnf_decription', 'w') as f:
  f.writelines(line + '\n' for line in bnf_description)
```
